# Basic plotting

```python
import pandas as pd
```

```python
oo = pd.read_csv('./data/olympics.csv',skiprows=4)
oo.head()
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## Plot types

### What were the different sports in the first olympics? Plot them using different graphs.

```python
fo = oo[oo.Edition == 1896]
fo.head()
```

### Line plot

```python
fo.Sport.value_counts().plot(kind='line');
#semicolon helps in supressing the matplot line 
#default is line graph
```

### Bar plot

```python
fo.Sport.value_counts().plot(kind='bar');
```

### Horizontal bar plot

```python
fo.Sport.value_counts().plot(kind='barh');
```

### Pie chart

```python
fo.Sport.value_counts().plot(kind='pie');
```

## Plot colors

```python
fo.Sport.value_counts().plot(color='plum')
```

```python
fo.Sport.value_counts().plot(kind='bar',color='maroon');
```

## figsize()

```python
fo.Sport.value_counts().plot(figsize=(10,3));
```

## Colormaps

```python
fo.Sport.value_counts().plot(kind='pie',colormap='Paired');
```

```python
fo.Sport.value_counts().plot(kind='pie',color='red');
```

# Seaborn basic plotting

```python
import seaborn as sns
```

### How many medals have been won by men and women in the history of the Olympics. How many gold, silver and bronze medals were won for each gender?

```python
sns.countplot(x='Medal',data=oo, hue='Gender');
```

```python
# Plot the numder of medals achieved by the chinese team in beijing 2008 using matplotlib and seaborn
b = oo[(oo.Edition == 2008) & (oo.NOC == 'CHN')]

sns.countplot(x='Gender', data=b, palette='bwr');
```

```python
#matplotlib
b.Gender.value_counts().plot(kind='bar', colormap='twilight');
```

```python
#plot the numder of gold silver and bronze medals
```

```python
sns.countplot(x='Medal', data=b, hue='Gender', palette='bwr', order=['Gold', 'Silver', 'Bronze']);
```
